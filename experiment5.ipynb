```python
import numpy as np
import matplotlib.pyplot as plt
X = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 3, 4, 5])
n = len(y)
theta0 = 0.0
theta1 = 0.0
def hypothesis(theta0, theta1, x):
    return theta0 + theta1 * x
def cost_function(theta0, theta1, x, y):
    total_cost = 0.0
    for i in range(n):
        total_cost += (hypothesis(theta0, theta1, x[i]) - y[i]) ** 2
    return total_cost / (2 * n)
theta1_values = np.linspace(-13, 11, 100)
cost_values = []
for theta1_val in theta1_values:
    cost = cost_function(theta0, theta1_val, X, y)
    cost_values.append(cost)
plt.figure(figsize=(8, 6))
plt.plot(theta1_values, cost_values, label='Cost Function')
plt.xlabel('theta1')
plt.ylabel('Cost')
plt.title('Cost Function for Linear Regression')
plt.legend()
plt.grid(True)
plt.savefig('cost_function_plot.png')
plt.show()
initial_cost = cost_function(theta0, theta1, X, y)
print(f"Initial cost (MSE) with theta0={theta0} and theta1={theta1}: {initial_cost}")
theta0_exp = 0.2
theta1_exp = 0.93
final_cost = cost_function(theta0_exp, theta1_exp, X, y)
print(f"Final cost (MSE) with theta0={theta0_exp} and theta1={theta1_exp}: {final_cost}")
```
